```python
# Create a list for numerical columns
num_cols = ['CSUSHPISA','unemp', 'Interest', 'CPI',
       'const_price_index', 'working_pop', 'Houses', 'per_capita_GDP',
       'Num_household', 'Income', 'Subsidy', 'Urban_pop', 'Old_pop']
```

```python
# Univariate Analysis using distplot and checking skewness of each variable
plt.figure(figsize=(25,25))
r=5
c=3
repeator =1
for i in num_cols:
    plt.subplot(r,c,repeator)
    sns.distplot(data.loc[:,i])
    plt.title('Skewness: %.2f'%data[i].skew())
    repeator+=1
plt.show()
plt.tight_layout()
```

```python
# Checking outlier using boxplot visulization
plt.figure(figsize=(20,20))
r=5
c=3
repeator =1
for i in num_cols:
    plt.subplot(r,c,repeator)
    sns.boxplot(data.loc[:,i])
    repeator+=1
plt.show()
plt.tight_layout()
```

```python
# Removing outlier using IQR method
for i in num_cols:
    q1,q3 = data[i].quantile([0.25,0.75])
    
    iqr = q3-q1
    
    ll = q1 - 1.5*iqr
    ul = q3 + 1.5*iqr
    
    data[i] = data[i].loc[~((data[i]<ll)|(data[i]>ul))]
    
```

```python
plt.figure(figsize=(20,20))
r=5
c=3
repeator =1
for i in num_cols:
    plt.subplot(r,c,repeator)
    sns.boxplot(data.loc[:,i])
    repeator+=1
plt.show()
plt.tight_layout()
```

```python
# Checking null values and treat them accordingly

data.isnull().sum()
```

```python
# Impute null values with mean of the column

mean1 = data['CSUSHPISA'].mean()
data['CSUSHPISA'] = data['CSUSHPISA'].fillna(mean1)

mean2 = data['unemp'].mean()
data['unemp'] = data['unemp'].fillna(mean2)

mean3 = data['const_price_index'].mean()
data['const_price_index'] = data['const_price_index'].fillna(mean3)
```

```python
# drop year, month and date columns as they have no more use
data.drop(columns = ['Year', 'Month', 'Date'], inplace =True)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split

x = data.drop('CSUSHPISA',axis =1)
y = data['CSUSHPISA']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=40)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

# Model Building with Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
model = lr.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
# Check r2 score of the model

from sklearn.metrics import r2_score

score = r2_score(y_test, y_pred)

print(f'The score of model is {score }')
```

```python
# check for other test parameters MSE, RMSE, MAE, MAPE

from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
```

```python
mse= mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test,y_pred)
mape = mean_absolute_percentage_error(y_test,y_pred)
```

```python
print(f'MSE : {mse}')
print(f'RMSE : {rmse}')
print(f'MAE : {mae}')
print(f'MAPE : {mape}')
```

```python
coefs = model.coef_
cols = x.columns
for i in range(len(coefs)):
    print(f"The coefficient for {cols[i]} is {coefs[i]}")
```

The coefficients of the variables are not giving us better correlation coefficients with the target variable for some variable.
The Consumer Price Index is effect not showing good coefficient as with increase in every unit of CPI House Price must also increase but indirectly due to other factors like interest rates, Inflation, increase in supply and demand.

```python
model.intercept_
```

An intercept value of 20 in the context of a linear regression model means that when all independent variables are set to zero, the predicted value of the dependent variable is 20. The intercept is essentially the starting point or baseline prediction of the model.

The model can perform much better if we have more features to describe in the dataset and sufficient data to train
